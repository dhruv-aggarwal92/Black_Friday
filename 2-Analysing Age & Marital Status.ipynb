```python
import pandas as pd
```

```python
data = pd.read_csv('Black Friday.csv')
data.head()
```

```python
del data["Product_Category_2"]
del data["Product_Category_3"]
```

```python
data.groupby('Age').size().plot(kind='bar',y='Purcahse')              #single col
```

```python
lst = []
for i in data['Age'].unique():
    lst.append([i,data[data['Age']==i]['Product_ID'].nunique()])
lst = pd.DataFrame(lst, columns=['Age','Unique prod'])
lst.plot.bar(x='Age',y='Unique prod')
lst
```

```python
data.groupby('Age').nunique()['Product_ID']
```

```python
data.groupby('Age').sum()['Purchase'].plot(kind='bar')
```

```python
data.groupby('Age').mean()['Purchase'].plot(kind='pie',autopct='%.1f')
```

```python
data.groupby('Marital_Status').size().plot(kind='pie',autopct='%.1f')
```
