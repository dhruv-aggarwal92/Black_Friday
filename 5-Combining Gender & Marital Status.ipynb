```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Black Friday.csv')
data.head()
```

```python
del data["Product_Category_2"]
del data["Product_Category_3"]
```

```python
l = [];
for i in range(len(data)):
    l.append(data['Gender'][i]+"_"+str(data['Marital_Status'][i]))
data['martialgender']=l
```

```python
data
```

```python
sns.countplot(x='martialgender',hue='City_Category',data=data)
```

```python
sns.countplot(x='martialgender',hue='Age',data=data)
```
